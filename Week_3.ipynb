```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# define an array x of atleast 100 points with a separation of at least 1
x = np.arange(1,101,1)
```

```python
# now choose any 2 values for A, B and generate an array y_true such that y_true = Ax + B for all x defined above
#let A= 1 and B= 2
y_true = np.arange(1,101,1)
i=0 
while i<100 :
  y_true[i] = x[i]+2
  i+=1
```

```python
# Add Gaussian Noise which can vary from upto -20 to 20 to y_true to get our data
min_noise = -20
max_noise = 20
sd_noise = (max_noise - min_noise) / 6
noise = np.random.normal(0, sd_noise, 100)
final_noise = (max_noise - min_noise) * (noise - noise.min()) / (noise.max() - noise.min()) + min_noise
data = np.arange(1,101,1)
i=0
while i <100 :
  data[i] = y_true[i] + final_noise[i]
  i+=1
```

```python
# print the data, y_true and x
print (x)
print (y_true)
print (data)
```

```python
# plot scatterplot between data and x
# also plot y_true vs x on the same plot
plt.scatter(data , x , color = 'b')
plt.scatter(y_true , x , color = 'r')
```

```python
# write a function definition to calculate loss of linear Regression
# I will be calculating MSE which measures the average squared difference between the predicted and true values
def calc_loss(y_true, y_pred):
  y_true = np.array(y_true)
  y_pred = np.array(y_pred)
  loss = np.mean((y_true -y_pred)**2)
  return loss
```

```python
# write a function definition to calculate derivative of cost function of linear regression with respect to parameter A
# write a function definition to calculate derivative of cost function of linear regression with respect to parameter B
#for the above case
def der_wrt_A(x , y_pred , data) :
  x = np.array(x)
  y_pred = np.array(y_pred)
  data = np.array(data)
  der_A = np.mean((y_pred - data)*x)
  return der_A
def der_wrt_B(y_pred , data) :
  y_pred = np.array(y_pred)
  data = np.array(data)
  der_B = np.mean((y_pred - data))
  return der_B
```

```python
# define A = 0 and B = 0 randomly
# also define 2 lists to store values of A and B at various iterations
A=0
B=0
A_values = [0]
B_values = [0]
```

```python
# compute the model array based on initial values of A and B
y_pred = np.array(A*x + B)
# Also print the cost function for the data before training
print(calc_loss(data, y_pred))
```

```python
# Define the Learning rate, alpha randomly. You will have to modify it during training
# write the main training loop and train for atleast 2000 iterations
A=0
B=0
A_values = [0]
B_values = [0]
i=0
while i<2000 :
  y_pred = np.array(A*x + B)
  print(calc_loss(data, y_pred))# print the cost at each iteration while training
  der_cost_A = der_wrt_A(x , y_pred , data)
  der_cost_B = der_wrt_B(y_pred , data)
  A = (A - 0.00001*der_cost_A)
  B = (B - 1*der_cost_B) 
  if i%400 ==0 : # save the values of A and B at atleast 5 different iterations spaced by atleast 50 into the lists defined earlier
    A_values.append(A)
    B_values.append(B)
  i+=1
print(A_values)
print(B_values)
# print the two lists for A and B and see how those values change with number of iterations
#About 1200 iterations were sufficient with the following learning rates i took by hit and trial randomly
```

```python
# generate different models for A and B you stored in the lists while training
A_values_int = [float(x) for x in A_values]
B_values_int = [float(x) for x in B_values]
plot_1 = np.arange(1,101,1)
i=0 
while i<100 :
  plot_1[i] = A_values_int[0]*x[i]+B_values_int[0]
  i+=1
plot_2 = np.arange(1,101,1)
i=0 
while i<100 :
  plot_2[i] = int(int(A_values_int[1])*x[i]+int(B_values_int[1]))
  i+=1
plot_3 = np.arange(1,101,1)
i=0 
while i<100 :
  plot_3[i] = A_values_int[2]*x[i]+B_values_int[2]
  i+=1
plot_4 = np.arange(1,101,1)
i=0 
while i<100 :
  plot_4[i] = A_values_int[3]*x[i]+B_values_int[3]
  i+=1
plot_5 = np.arange(1,101,1)
i=0 
while i<100 :
  plot_5[i] = A_values_int[4]*x[i]+B_values_int[4]
  i+=1
plot_6 = np.arange(1,101,1)
i=0 
while i<100 :
  plot_6[i] = A_values_int[5]*x[i]+B_values_int[5]
  i+=1
# plot the different models you defined earlier and also y_true vs x on the same plot
# use different colours, though the lines may not be visible distinctly which is ok
plt.scatter(plot_1 , x , color = 'b')
plt.scatter(plot_2 , x , color = 'r')
plt.scatter(plot_3 , x , color = 'y')
plt.scatter(plot_4 , x , color = 'black')
plt.scatter(plot_5 , x , color = 'brown')
plt.scatter(plot_6 , x , color = 'pink')
plt.scatter(y_true , x , color = 'purple')

```
