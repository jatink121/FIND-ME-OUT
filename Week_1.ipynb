## ***Question 1***

```python
a= int(input("Give the number n->"))
arr=[];
i=0;
while i<int(a):
    arr.append(int(input("Give the "+str((i+1))+"th entry of array->")))
    i+=1;
i=0;
while i<a:
    j=i+1
    m=0
    while j<a:
        if arr[i]>=arr[j] :
            m+=1
        j+=1
    if m==(a-i-1):
        print(arr[i])
    i+=1
```

## ***Question 2***

```python
a= int(input("Give the number n->"))
arr=[];
i=0;
while i<int(a):
    arr.append(int(input("Give the "+str((i+1))+"th entry of array->")))
    i+=1;
i=1;sum=0;max_1=0;max_2=0
while i<a-1:
    j=0
    while j<i :
        if arr[j]> arr[i] :
            if arr[j]> max_1:
                max_1 = arr[j]
        j+=1 
    while j<a :
        if arr[j]> arr[i] :
            if arr[j]> max_2:
                max_2 = arr[j]
        j+=1 
    if max_2>max_1 :
        sum+=(max_1-arr[i])
    if max_1>max_2 :
        sum+=(max_2-arr[i])
    i+=1
print (sum)
```
