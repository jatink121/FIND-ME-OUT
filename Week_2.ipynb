Q-Q plots : Quantile -Quantile plots 
These are the plots that are plotted betwwen 2 dta sets of data points to compare them with each other.
Suppose we have a set of data points and then we find the sample quantilesof this data and plot them against the tepretical quantiles where we assue the data points to be normally distributed for example.
if the the graph plotted lies across a straight line we conclude that the set of data pointss is normally distributed.
What i have done here is that for given data points , i have calculated 1,2,3,...100 quantile.
ALso i assumed 150 normally distributed data points in the same range then i plotted the graph of these 2 .
If it turns to be a straight line we can conclude that the points are normally distributed.

```python
h_1.sort()

Quantile_2 = np.arange(1.0,101.0,1.0)
i_quant_2=1
while i_quant_2<100:
  a_index = int(i_quant_2*1.5 - (i_quant_2*1.5)%1)
  Quantile_2[i_quant_2-1]= h_1[a_index]
  i_quant_2+=1
  Quantile_2[99]=  h_1[149]
h_th_2=np.random.normal(4,8,150)
h_th_2.sort()
Quantile_2_th = np.arange(1.0,101.0,1.0)
i_quant_2_th=1
while i_quant_2_th<100:
  a_index_th = int(i_quant_2_th*1.5 - (i_quant_2_th*1.5)%1)
  Quantile_2_th[i_quant_2_th-1]= h_th_2[a_index_th]
  i_quant_2_th+=1
  Quantile_2_th[99]=  h_th_2[149]
plt.scatter(Quantile_2 , Quantile_2_th)
```
